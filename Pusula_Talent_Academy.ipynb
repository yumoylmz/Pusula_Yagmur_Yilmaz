```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
plt.style.use('ggplot')
```

```python
df = pd.read_excel("physical_med_reh.xlsx")
```

```python
df
```

```python
df.shape
```

```python
df.info()
```

```python
df.isnull().sum()
```

```python
categorical_cols = ['KanGrubu', 'Cinsiyet', 'Uyruk', 'Bolum',
                    'KronikHastalik', 'Alerji', 'Tanilar','TedaviAdi','UygulamaYerleri']

for col in categorical_cols:
    df[col] = df[col].astype(str)
    df[col] = df[col].str.strip()
    df[col] = df[col].str.lower()
```

```python
df['TedaviAdi'].unique()
```

```python
df['KanGrubu'].value_counts()
```

```python
df['TedaviAdi'].value_counts()
```

```python
df['Bolum'].unique()
```

```python
from sklearn.impute import SimpleImputer

cat_cols = ["Cinsiyet","KanGrubu","Bolum","Alerji","Tanilar","UygulamaYerleri","KronikHastalik"]

for col in cat_cols:

    df[col] = df[col].astype(str).str.strip()


    df[col] = df[col].replace(['<NA>','nan','NaN','NAN',''], np.nan)

    df[col] = df[col].astype('object')

cat_imputer = SimpleImputer(strategy="most_frequent")
df[cat_cols] = cat_imputer.fit_transform(df[cat_cols])
```

```python
df.isnull().sum()
```

```python
df['Alerji'].unique()
```

```python
df.duplicated().sum()
```

```python
df.nunique()
```

```python
df = df.drop(columns=["HastaNo"])
```

```python
df["TedaviSuresi"] = df["TedaviSuresi"].astype(str).str.extract(r"(\d+)").astype(float)
df["UygulamaSuresi"] = df["UygulamaSuresi"].astype(str).str.extract(r"(\d+)").astype(float)
```

```python
df
```

```python
df.describe()
```

```python
# 2. Target dağılımı
sns.histplot(df['TedaviSuresi'], kde=True)
plt.show()
```

```python
df['Uyruk'].value_counts()
```

```python
df['Uyruk'].value_counts().plot(kind='bar')
plt.title("Uyruk Dağılımı")
plt.xticks(rotation=45)
plt.show()
```

```python
plt.figure(figsize=(8,5))
sns.boxplot(x='Uyruk', y='TedaviSuresi', data=df)
plt.title("Uyruklara Göre Tedavi Süresi Dağılımı")
plt.xticks(rotation=45)
plt.show()
```

```python
country_age = df.groupby('Uyruk')['Yas'].mean().sort_values(ascending=False)
country_age.plot(kind='bar', figsize=(10,5))
plt.title("Ülkelere Göre Ortalama Hasta Yaşı")
plt.ylabel("Ortalama Yaş")
plt.xlabel("Ülke")
plt.xticks(rotation=45)
plt.show()
```

```python
sns.countplot(x='Cinsiyet', hue='Cinsiyet', data=df, palette='Set2')
plt.title('cinsiyet ve tedavi süresi')
plt.xlabel('Cinsiyet')
plt.ylabel('Tedavi süresi')
plt.xticks(rotation=60,ha='right')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
plt.scatter(df['Yas'], df['TedaviSuresi'], alpha=0.6, color='y')
plt.title('Yas oranına göre tedavi süresi')
plt.xlabel('yas')
plt.ylabel('Tedavi süresi')
plt.grid(True)
plt.show()
```

```python
plt.figure(figsize=(8,5))
sns.histplot(df["TedaviSuresi"], color="blue", kde=True, label="Tedavi Süresi")
sns.histplot(df["UygulamaSuresi"], color="red", kde=True, label="Uygulama Süresi")
plt.legend()
plt.title("Tedavi vs Uygulama Süresi Dağılımı")
plt.show()
```

```python
top_n = 20
treatment_count = df['TedaviAdi'].value_counts()

treatment_count.head(top_n).plot(kind='bar', figsize=(10,6))
plt.title("En Çok Görülen İlk 20 Tedavi Türü")
plt.xlabel("Tedavi Adı")
plt.ylabel("Hasta Sayısı")
plt.xticks(rotation=45, ha="right")
plt.show()
```

```python
dept_treatment = df.groupby("Bolum")["TedaviSuresi"].mean().sort_values()
dept_treatment.plot(kind="line", marker="o", figsize=(10,5))
plt.title("Bölümlere Göre Ortalama Tedavi Süresi")
plt.xlabel("Bölüm")
plt.ylabel("Ortalama Tedavi Süresi")
plt.xticks(rotation=45, ha="right")
plt.show()
```

```python
cols = ['Yas', 'TedaviSuresi', 'UygulamaSuresi']
correlation_matrix = df[cols].corr()

plt.figure(figsize=(10, 8))
sns.heatmap(correlation_matrix, annot=True, cmap="RdYlGn", linecolor='white', linewidths=0.3)
plt.title("Yaş, Tedavi Süresi ve Uygulama Süresi Korelasyonu", fontsize=16)
plt.show()
```
